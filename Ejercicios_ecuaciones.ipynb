```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Ejercicio 1
#a
a = np.array([
    [5,2],
    [-3,3]
])
b = np.array([1,5])

x_solucion = np.linalg.solve(a,b)
print("Solucion:", x_solucion)


x_values = np.linspace(-2, 3, 100)

y1_values = (1 - 5 * x_values) / 2

y2_values = (5 + 3 * x_values) / 3

plt.figure(figsize=(8,6))
plt.plot(x_values, y1_values, label='5x + 2y = 1')
plt.plot(x_values, y2_values, label='-3x + 3y = 5')
plt.plot(x_solucion[0], x_solucion[1], 'ro', label='Solucion')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Ejercicio 1')
plt.legend()
plt.grid(True)
plt.show()
plt.close()
```
